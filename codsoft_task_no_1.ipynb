```python
import kagglehub as kh

# Download latest version
path = kh.dataset_download("hijest/genre-classification-dataset-imdb")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
import re
```

# Data Exploration using Pandas

```python
train_data=pd.read_csv("/kaggle/input/genre-classification-dataset-imdb/Genre Classification Dataset/train_data.txt",sep=':::',names=["movie_name","genre","summary"])
test_data=pd.read_csv("/kaggle/input/genre-classification-dataset-imdb/Genre Classification Dataset/test_data.txt",sep=':::',names=["movie_name","summary"])
train_data.head()
```

```python
train_data.describe()
```

# Data cleaning

```python
def clean_text(text):
  text=text.lower()
  text=re.sub(r"[^a-z0-9\s]","",text)
  text=re.sub(R"\s+"," ",text).strip()
  return text
```

```python
print(train_data.columns)
```

```python
#For training data
train_data['clean_plot']=train_data["summary"].apply(clean_text)
train_data[['movie_name', 'genre', 'summary','clean_plot']].head(10)
```

```python
# For testing data
test_data['clean_plot']=test_data['summary'].apply(clean_text)
test_data[['movie_name','summary','clean_plot']].head(10)
```

## Feature Extraction (Conversion of text to numbers in input data)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
obj=TfidfVectorizer(stop_words="english",max_features=5000)
X_train=obj.fit_transform(train_data['clean_plot'])

X_test=obj.transform(test_data['clean_plot'])
```

### Label Encoding (Conversion of output data from text to numbers)

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lab_encod=LabelEncoder()
Y_train=lab_encod.fit_transform(train_data['genre'])
```

# Model Training

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression(class_weight='balanced',max_iter=1000)
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
predicted_genres=lab_encod.inverse_transform(y_pred)
test_data['predicted_genres']=predicted_genres
test_data.head()
```

# Prediction

```python
test_data[['movie_name','summary','predicted_genres']].to_csv("predicted_genre.csv",index=False)
```

# Testing accuracy in the model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
```

```python
x_train_split,x_test_split,y_train_split,y_test_split=train_test_split(X_train,Y_train,test_size=0.2,random_state=42)
model.fit(x_train_split,y_train_split)
y_pred_split=model.predict(x_test_split)
```

```python
acc=accuracy_score(y_test_split,y_pred_split)
print(f"Accuracy :{acc:.2f}")
```

```python
print("Classification Report:\n")
print(classification_report(y_test_split,y_pred_split,target_names=lab_encod.classes_))
```
